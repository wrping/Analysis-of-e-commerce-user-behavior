#### 3.1.1 销量前10的产品与产品类别

```python
#销量前10的产品
```

```python
top = df.groupby('product_id').agg(销售总量=('buy_cnt','sum')).reset_index()
        .sort_values('销售总量',ascending=False).reset_index(drop=True)
top.head(10)
```

#### 3.1.3 手机品牌表现较好的品牌

```python
#前面说销量最好的电子产品是手机，这里分析手机品牌下面表现最好的5个品牌
```

```python
brand_5 = df[df.category_code=='electronics.smartphone'].groupby('brand')
    .agg(销量=('buy_cnt','sum')).reset_index().sort_values('销量',ascending=False)
brand_5.reset_index(drop=True,inplace=True)
brand_5.head(5)
```

```python
#表现最好的5个品牌分别为三星、苹果、小米、华为和oppo，比较符合我们的认知
```

#### 3.1.4 三星手机对应的人群特征

```python
#通过以上分析可知，三星手机在所有的电子产品中非常占优势，不管销量还是销售额都属于佼佼者，因此分析三星手机所对应的人群特征
```

#### 1 年龄分布

```python
bins = [15,20,25,30,35,40,45,50]
labels = ['(15-20]岁','(20-25]岁','(25-30]岁','(30-35]岁','(35-40]岁','(40-45]岁','(45-50]岁']
df_age = df.copy()
df_age['age_bin'] = pd.cut(x=df.age,bins=bins,right=True,labels=labels)
```

```python
user_sanxing = df_age.loc[(df_age['brand']=='samsung') & (df_age['category_code']=='electronics.smartphone')]
user_sanxing.head()
```

```python
user_sanxing.groupby('age_bin').agg(销量=('buy_cnt','sum'))
```

```python
#年龄分布比较平均，仍然是35-40岁的人群较少
```

#### 2 城市分布

```python
local_brand = user_sanxing.groupby('local').agg(销量=('buy_cnt','sum')).sort_values('销量',ascending=False)
local_brand
```

#### 3 性别分布

```python
user_sanxing.groupby('sex').agg(销量=('buy_cnt','sum'))
```
